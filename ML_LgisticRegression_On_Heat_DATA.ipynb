```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('heart.csv')
```

```python
df.drop_duplicates(inplace=True)
```

```python
X= df.drop('target',axis=True)
y = df['target']
```

```python
print(X.shape)
y.shape
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression,LogisticRegressionCV
```

```python
logRef =LogisticRegression()
```

```python
logRef.fit(X_train,y_train)
```

```python
y_pred_train = logRef.predict(X_train)
```

```python
from sklearn.metrics import classification_report,accuracy_score,r2_score,confusion_matrix
```

```python
print(classification_report(y_train,y_pred_train))
```

```python
accuracy_score(y_train,y_pred_train)
```

```python
confusion_matrix(y_train,y_pred_train)
```

### TEST PREDICTION

```python
y_pred_test= logRef.predict(X_test)
```

```python
print(classification_report(y_test,y_pred_test))
```

```python
from sklearn.model_selection import cross_val_score,KFold
```

```python
num_fold= 5
kf = KFold(n_splits=num_fold,random_state=0,shuffle=True)
```

```python
cv = cross_val_score(logRef,X,y,cv=kf)
```
